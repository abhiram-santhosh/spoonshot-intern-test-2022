```python
# !pip install sentence-transformers # only needed first time
from sentence_transformers import SentenceTransformer, util

# using pre-trained model
model = SentenceTransformer('all-MiniLM-L6-v2')

# embedding the queries and article
query_embeddings = model.encode(query_list, convert_to_tensor=True)
article_embedding = model.encode(article, convert_to_tensor=True)

# finding similarity of each query with the article
cosine_scores = util.cos_sim(query_embeddings,article_embedding)
query_scores = dict(zip(query_list,cosine_scores))

# displaying queries in decreasing order of similarity
relevant_queries = dict(sorted(query_scores.items(),key=lambda item: item[1],reverse=True))
list(relevant_queries.keys())
```
